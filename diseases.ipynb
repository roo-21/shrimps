```python
diseases=dict()
```

```python
for index,row in df.iterrows():
    key=row['Disease']
    values=set(df.columns[row==1])
    diseases[key]=values
print(diseases)
```

```python
with open('/content/drive/MyDrive/s3 mini p/diseases.pkl','wb') as file:
    pickle.dump(diseases,file)
```
